- requests로 url 호출

```python
import requests
url = 'https://ko.wikipedia.org/wiki/%EC%9D%B8%EA%B3%B5%EC%A7%80%EB%8A%A5'
res = requests.get(url)
res
```

200이면 정상적으로 동작한 것

- BeautifulSoup으로 HTML 문서 해석

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(res.text, 'html.parser')
soup
```

- 필요 부분 선택 : CSS 선택자 사용 = select

```python
soup.select_one('#firstHeading') # select 가장 첫번째 것 불러오기
```

```python
soup.select_one('#mw-content-text > div.mw-parser-output')
```

```python
soup.select('#firstHeading')
```

```python
soup.select('#firstHeading')[0].get_text()
```

- h2에 해당하는 내용 모두 찾기 : 태그 기반 검색 = find

```python
soup.find('h2')
```

```python
soup.find_all('h2')
```

```python
[i.get_text() for i in soup.find_all('h2')]
```
